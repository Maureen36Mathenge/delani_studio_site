```
#loading dataset
df2 = pd.read_csv("Telcom_dataset.csv")
print(df2)
df3 = pd.read_csv("Telcom_dataset2.csv")
print(df3)
df4 = pd.read_csv("Telcom_dataset3.csv")
print(df4)

```

```
#rename certain columns so that they can merge
df02 = df2.rename(columns={"PRODUTC":"PRODUCT","DATETIME":" DATE_TIME"})
df02
```

```
df04 = df4.rename(columns={"CELLID" : "CELL_ID","SIET_ID" : "SITE_ID"})
df04
```

```
#merge the datasets
df5 = df04.merge(df02, how='left').merge(df3, how='left')
df5
```

```
df5.info()
```

```
df5.describe()
#On average billing price on customers is 25(round off)
```

```
#Check if the columns are correlated
df5.corr(method ='kendall')
#there is a strong correlation between the columns in the dataset
```

```
#drop columns
df05 = df5.drop(["COUNTRY_A" , "COUNTRY_B"] , axis = 1 , inplace = False) 
df05
```

```
#joing the cell_geo data set and the merged dataset
df11 = pd.concat([df05, df1] , axis = 1 )
df11
```

```
df11.info()
```

```
#fill values with NaN with o
df11.fillna(0)

```

```
#group to get a vivid picture
df11.groupby(["VILLES","PRODUCT"])["VALUE"].sum().head(40)
#cities with the most cell_id have many more calls taking place.
#ABOBO which has the 3rd most cell_id had 2765 calls

#df12.sort_values(by = "VILLES", ascending = False).head(10)
```
